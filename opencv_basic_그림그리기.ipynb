```python
import cv2                    # Ïù¥ÎØ∏ÏßÄ Ï≤òÎ¶¨Ïö© OpenCV
import numpy as np            # Î∞∞Ïó¥ ÏÉùÏÑ±ÏùÑ ÏúÑÌïú numpy
import matplotlib.pyplot as plt  # ÏãúÍ∞ÅÌôîÎ•º ÏúÑÌïú matplotlib
from google.colab import files  # ColabÏóêÏÑú ÌååÏùº ÏóÖÎ°úÎìúÏö© (ÌòÑÏû¨ ÏΩîÎìúÏóêÏÑúÎäî ÏÇ¨Ïö© Ïïà Îê®)
import io
from PIL import Image         # PillowÎäî Ïù¥ÎØ∏ÏßÄ Î∂àÎü¨Ïò§Í∏∞/Ï†ÄÏû•Ïóê ÏÇ¨Ïö© Í∞ÄÎä• (ÌòÑÏû¨ ÏΩîÎìúÏóêÏÑúÎäî ÏÇ¨Ïö© Ïïà Îê®)


print(f"üéâ OpenCV Î≤ÑÏ†Ñ: {cv2.__version__}")

# =============================================================================
# üìù ÌÖåÏä§Ìä∏ 1: Í∞ÑÎã®Ìïú Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ±ÌïòÍ∏∞
# =============================================================================

def test1_create_image():
    """Í∞ÑÎã®Ìïú Ïù¥ÎØ∏ÏßÄ ÎßåÎì§Ïñ¥Î≥¥Í∏∞"""
    print("\nüé® ÌÖåÏä§Ìä∏ 1: Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ±ÌïòÍ∏∞")

    # 1. Îπà Ï∫îÎ≤ÑÏä§ ÎßåÎì§Í∏∞ (Í≤ÄÏùÄÏÉâ)
    black_img = np.zeros((200, 300, 3), dtype=np.uint8)

    # 2. Ìù∞ÏÉâ Ï∫îÎ≤ÑÏä§ ÎßåÎì§Í∏∞
    white_img = np.ones((200, 300, 3), dtype=np.uint8) * 255

    # 3. Îπ®Í∞ÑÏÉâ Ï∫îÎ≤ÑÏä§ ÎßåÎì§Í∏∞
    red_img = np.zeros((200, 300, 3), dtype=np.uint8)
    red_img[:, :, 2] = 255  # BGRÏóêÏÑú RÏ±ÑÎÑê

    # Í≤∞Í≥º Ï∂úÎ†•
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 3, 1)
    plt.imshow(black_img)
    plt.title('Black Image')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(white_img)
    plt.title('White Image')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(cv2.cvtColor(red_img, cv2.COLOR_BGR2RGB))
    plt.title('Red Image')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
# Ìï®Ïàò Ïã§Ìñâ ÏΩîÎìú Ï∂îÍ∞Ä
if __name__ == "__main__":
    test1_create_image()
```

```python
# =============================================================================
# üìù ÌÖåÏä§Ìä∏ 2: ÎèÑÌòï Í∑∏Î¶¨Í∏∞
# =============================================================================

def test2_draw_shapes():
    """ÎèÑÌòï Í∑∏Î¶¨Í∏∞ Ïó∞Ïäµ"""
    print("\nüî∫ ÌÖåÏä§Ìä∏ 2: ÎèÑÌòï Í∑∏Î¶¨Í∏∞")

    # Ï∫îÎ≤ÑÏä§ ÏÉùÏÑ±
    canvas = np.ones((400, 600, 3), dtype=np.uint8) * 255

    # 1. ÏÇ¨Í∞ÅÌòï Í∑∏Î¶¨Í∏∞
    cv2.rectangle(canvas, (50, 50), (200, 150), (255, 0, 0), 3)  # ÌååÎûÄÏÉâ ÌÖåÎëêÎ¶¨
    cv2.rectangle(canvas, (250, 50), (400, 150), (0, 255, 0), -1)  # Ï¥àÎ°ùÏÉâ Ï±ÑÏö∞Í∏∞

    # 2. Ïõê Í∑∏Î¶¨Í∏∞
    cv2.circle(canvas, (125, 250), 50, (0, 0, 255), 3)  # Îπ®Í∞ÑÏÉâ ÌÖåÎëêÎ¶¨
    cv2.circle(canvas, (325, 250), 50, (255, 255, 0), -1)  # ÎÖ∏ÎûÄÏÉâ Ï±ÑÏö∞Í∏∞

    # 3. ÏßÅÏÑ† Í∑∏Î¶¨Í∏∞
    cv2.line(canvas, (450, 50), (550, 150), (255, 0, 255), 5)  # Î≥¥ÎùºÏÉâ ÏÑ†

    # 4. ÌÖçÏä§Ìä∏ Ïì∞Í∏∞
    cv2.putText(canvas, 'Hello OpenCV!', (200, 350),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)

    # Í≤∞Í≥º Ï∂úÎ†•
    plt.figure(figsize=(10, 6))
    plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
    plt.title('Shape Drawing Test')
    plt.axis('off')
    plt.show()

# Ìï®Ïàò Ïã§Ìñâ ÏΩîÎìú Ï∂îÍ∞Ä
if __name__ == "__main__":
    test2_draw_shapes()
```

```python
# =============================================================================
# üìù ÌÖåÏä§Ìä∏ 3: ÏÉâÏÉÅ Î≥ÄÌôò
# =============================================================================

def test3_color_conversion():
    """ÏÉâÏÉÅ Î≥ÄÌôò ÌÖåÏä§Ìä∏"""
    print("\nüåà ÌÖåÏä§Ìä∏ 3: ÏÉâÏÉÅ Î≥ÄÌôò")

    # Ïª¨Îü¨ Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ± (Î¨¥ÏßÄÍ∞ú Ìå®ÌÑ¥)
    img = np.zeros((200, 600, 3), dtype=np.uint8)

    # Î¨¥ÏßÄÍ∞ú ÏÉâÏÉÅ ÎßåÎì§Í∏∞
    colors = [
        (255, 0, 0),    # Îπ®Í∞ï
        (255, 127, 0),  # Ï£ºÌô©
        (255, 255, 0),  # ÎÖ∏Îûë
        (0, 255, 0),    # Ï¥àÎ°ù
        (0, 0, 255),    # ÌååÎûë
        (75, 0, 130),   # ÎÇ®ÏÉâ
        (148, 0, 211)   # Î≥¥Îùº
    ]

    # Í∞Å ÏÉâÏÉÅÏùÑ ÏÑ∏Î°ú Îù†Î°ú ÎßåÎì§Í∏∞
    strip_width = 600 // len(colors)
    for i, color in enumerate(colors):
        x_start = i * strip_width
        x_end = (i + 1) * strip_width
        img[:, x_start:x_end] = color

    # Îã§ÏñëÌïú ÏÉâÍ≥µÍ∞ÑÏúºÎ°ú Î≥ÄÌôò
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

    # Í≤∞Í≥º Ï∂úÎ†•
    plt.figure(figsize=(15, 10))

    plt.subplot(2, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Original (RGB)')
    plt.axis('off')

    plt.subplot(2, 2, 2)
    plt.imshow(gray, cmap='gray')
    plt.title('Grayscale')
    plt.axis('off')

    plt.subplot(2, 2, 3)
    plt.imshow(hsv)
    plt.title('HSV')
    plt.axis('off')

    plt.subplot(2, 2, 4)
    plt.imshow(lab)
    plt.title('LAB')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

# Ìï®Ïàò Ïã§Ìñâ ÏΩîÎìú Ï∂îÍ∞Ä
if __name__ == "__main__":
    test3_color_conversion()
```

```python
# =============================================================================
# üìù ÌÖåÏä§Ìä∏ 4: Ïù¥ÎØ∏ÏßÄ ÌïÑÌÑ∞ÎßÅ
# =============================================================================

def test4_image_filtering():
    """Ïù¥ÎØ∏ÏßÄ ÌïÑÌÑ∞ÎßÅ ÌÖåÏä§Ìä∏"""
    print("\nüîç ÌÖåÏä§Ìä∏ 4: Ïù¥ÎØ∏ÏßÄ ÌïÑÌÑ∞ÎßÅ")

    # ÎÖ∏Ïù¥Ï¶àÍ∞Ä ÏûàÎäî Ïù¥ÎØ∏ÏßÄ ÏÉùÏÑ±
    img = np.ones((300, 300, 3), dtype=np.uint8) * 128  # ÌöåÏÉâ Î∞∞Í≤Ω

    # Ïõê Í∑∏Î¶¨Í∏∞
    cv2.circle(img, (150, 150), 80, (255, 255, 255), -1)

    # ÎÖ∏Ïù¥Ï¶à Ï∂îÍ∞Ä
    noise = np.random.randint(0, 50, img.shape, dtype=np.uint8)
    noisy_img = cv2.add(img, noise)

    # ÌùëÎ∞± Î≥ÄÌôò
    gray = cv2.cvtColor(noisy_img, cv2.COLOR_BGR2GRAY)

    # Îã§ÏñëÌïú ÌïÑÌÑ∞ Ï†ÅÏö©
    blurred = cv2.GaussianBlur(gray, (15, 15), 0)
    edges = cv2.Canny(gray, 50, 150)

    # Í≤∞Í≥º Ï∂úÎ†•
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 4, 1)
    plt.imshow(cv2.cvtColor(noisy_img, cv2.COLOR_BGR2RGB))
    plt.title('Original (with noise)')
    plt.axis('off')

    plt.subplot(1, 4, 2)
    plt.imshow(gray, cmap='gray')
    plt.title('Grayscale')
    plt.axis('off')

    plt.subplot(1, 4, 3)
    plt.imshow(blurred, cmap='gray')
    plt.title('Gaussian Blur')
    plt.axis('off')

    plt.subplot(1, 4, 4)
    plt.imshow(edges, cmap='gray')
    plt.title('Edge Detection')
    plt.axis('off')

    plt.tight_layout()
    plt.show()

# Ìï®Ïàò Ïã§Ìñâ ÏΩîÎìú Ï∂îÍ∞Ä
if __name__ == "__main__":
    test4_image_filtering()
```

```python
# =============================================================================
# üìù ÌÖåÏä§Ìä∏ 5: Ïù¥ÎØ∏ÏßÄ ÏóÖÎ°úÎìú Î∞è Ï≤òÎ¶¨ (ÏÑ†ÌÉùÏÇ¨Ìï≠)
# =============================================================================

def test5_upload_and_process():
    """Ïù¥ÎØ∏ÏßÄ ÏóÖÎ°úÎìúÌïòÍ≥† Ï≤òÎ¶¨ÌïòÍ∏∞ (ÏÑ†ÌÉùÏÇ¨Ìï≠)"""
    print("\nüìÅ ÌÖåÏä§Ìä∏ 5: Ïù¥ÎØ∏ÏßÄ ÏóÖÎ°úÎìú Î∞è Ï≤òÎ¶¨")
    print("Ïù¥ÎØ∏ÏßÄ ÌååÏùºÏùÑ ÏóÖÎ°úÎìúÌïòÏÑ∏Ïöî...")

    try:
        # ÌååÏùº ÏóÖÎ°úÎìú
        uploaded = files.upload()

        if uploaded:
            filename = list(uploaded.keys())[0]

            # Ïù¥ÎØ∏ÏßÄ ÏùΩÍ∏∞
            image_data = uploaded[filename]
            img = Image.open(io.BytesIO(image_data))
            img_array = np.array(img)

            # RGB ‚Üí BGR Î≥ÄÌôò (OpenCVÏö©)
            if len(img_array.shape) == 3:
                img_bgr = cv2.cvtColor(img_array, cv2.COLOR_RGB2BGR)
            else:
                img_bgr = img_array

            # ÌÅ¨Í∏∞ Ï°∞Ï†ï (ÎÑàÎ¨¥ ÌÅ∞ Ïù¥ÎØ∏ÏßÄ Î∞©ÏßÄ)
            height, width = img_bgr.shape[:2]
            if width > 800:
                scale = 800 / width
                new_width = 800
                new_height = int(height * scale)
                img_bgr = cv2.resize(img_bgr, (new_width, new_height))

            # Îã§ÏñëÌïú Ï≤òÎ¶¨ Ï†ÅÏö©
            gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
            blurred = cv2.GaussianBlur(img_bgr, (21, 21), 0)
            edges = cv2.Canny(gray, 100, 200)

            # Í≤∞Í≥º Ï∂úÎ†•
            plt.figure(figsize=(20, 5))

            plt.subplot(1, 4, 1)
            plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
            plt.title('Original')
            plt.axis('off')

            plt.subplot(1, 4, 2)
            plt.imshow(gray, cmap='gray')
            plt.title('Grayscale')
            plt.axis('off')

            plt.subplot(1, 4, 3)
            plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
            plt.title('Blurred')
            plt.axis('off')

            plt.subplot(1, 4, 4)
            plt.imshow(edges, cmap='gray')
            plt.title('Edge Detection')
            plt.axis('off')

            plt.tight_layout()
            plt.show()

            print(f"‚úÖ Ïù¥ÎØ∏ÏßÄ Ï≤òÎ¶¨ ÏôÑÎ£å! ÌÅ¨Í∏∞: {img_bgr.shape}")

    except Exception as e:
        print(f"‚ùå ÏóÖÎ°úÎìú Ï§ë Ïò§Î•ò: {e}")
        print("Ïù¥ ÌÖåÏä§Ìä∏Îäî Í±¥ÎÑàÎõ∞Í≤†ÏäµÎãàÎã§.")

# Ìï®Ïàò Ïã§Ìñâ ÏΩîÎìú Ï∂îÍ∞Ä
if __name__ == "__main__":
   test5_upload_and_process()
```

```python
import cv2
import numpy as np

# Ïù¥ÎØ∏ÏßÄ ÏùΩÍ∏∞
img = cv2.imread('/mnt/data/a4dee0c1-0344-456e-b97d-945ac7976aad.png')

# Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞ Ï°∞Ï†ï (ÌïÑÏöîÏãú)
img_resized = cv2.resize(img, (640, 480))

# BGR -> HSV Î≥ÄÌôò (ÏÉâÏÉÅ Í≤ÄÏ∂úÏóê Ïú†Î¶¨)
hsv = cv2.cvtColor(img_resized, cv2.COLOR_BGR2HSV)

# Îπ®Í∞ÑÏÉâ Î≤îÏúÑ ÏÑ§Ï†ï (Ïòà: Ïã†Ìò∏Îì± Îπ®Í∞ï)
lower_red1 = np.array([0, 100, 100])
upper_red1 = np.array([10, 255, 255])
lower_red2 = np.array([160, 100, 100])
upper_red2 = np.array([179, 255, 255])

# Îπ®Í∞ÑÏÉâ ÎßàÏä§ÌÅ¨ ÏÉùÏÑ±
mask_red1 = cv2.inRange(hsv, lower_red1, upper_red1)
mask_red2 = cv2.inRange(hsv, lower_red2, upper_red2)
mask_red = cv2.bitwise_or(mask_red1, mask_red2)

# ÎßàÏä§ÌÅ¨Î•º ÌÜµÌï¥ Îπ®Í∞ÑÏÉâ ÏòÅÏó≠ Ï∂îÏ∂ú
red_result = cv2.bitwise_and(img_resized, img_resized, mask=mask_red)

# Ïú§Í≥ΩÏÑ† Ï∞æÍ∏∞ (Îπ®Í∞ÑÏÉâ ÏòÅÏó≠ Í≤ΩÍ≥Ñ)
contours, _ = cv2.findContours(mask_red, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours:
    area = cv2.contourArea(cnt)
    if area > 500:  # ÎÑàÎ¨¥ ÏûëÏùÄ ÏòÅÏó≠ÏùÄ Î¨¥Ïãú
        # Ïú§Í≥ΩÏÑ† Í∑ºÏÇ¨Ìôî(Î™®Ïñë Îã®ÏàúÌôî)
        approx = cv2.approxPolyDP(cnt, 0.02 * cv2.arcLength(cnt, True), True)

        # Î™®ÏñëÏù¥ ÏõêÏóê Í∞ÄÍπåÏö¥ÏßÄ ÌôïÏù∏ (Ïõê = Ïã†Ìò∏Îì± Î∂àÎπõ)
        (x, y), radius = cv2.minEnclosingCircle(cnt)
        center = (int(x), int(y))
        radius = int(radius)
        if len(approx) > 8:  # Îã§Í∞ÅÌòï Íº≠ÏßìÏ†êÏù¥ ÎßéÏúºÎ©¥ ÏõêÏúºÎ°ú ÌåêÎã®
            cv2.circle(img_resized, center, radius, (0, 255, 0), 2)
            cv2.putText(img_resized, "Red light", (center[0]-10, center[1]-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

cv2.imshow("Detected Red Light", img_resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
