```python
from ultralytics import YOLO
import cv2
from google.colab import files
import os
import time
import urllib.request

# â–¶ï¸ yolov12n.pt ë‹¤ìš´ë¡œë“œ URL (Ultralytics ê³µì‹ GitHubì—ì„œ ìµœì‹  í™•ì¸ í•„ìš”)
model_url = 'https://github.com/ultralytics/ultralytics/releases/download/v0.0.148/yolov12n.pt'
model_path = 'yolov12n.pt'

# ğŸ“¥ ëª¨ë¸ íŒŒì¼ì´ ì—†ìœ¼ë©´ ë‹¤ìš´ë¡œë“œ
#if not os.path.exists(model_path):
    #print(f"â¬‡ï¸ {model_path} íŒŒì¼ì´ ì—†ì–´ ë‹¤ìš´ë¡œë“œë¥¼ ì‹œì‘í•©ë‹ˆë‹¤...")
    #urllib.request.urlretrieve(model_url, model_path)
    #print(f"âœ… {model_path} ë‹¤ìš´ë¡œë“œ ì™„ë£Œ!")

# â–¶ï¸ YOLOv12 ëª¨ë¸ ë¡œë”©
model = YOLO(model_path)

# ğŸ“ ì˜ìƒ íŒŒì¼ ì—…ë¡œë“œ
print("\nğŸ“ ë¶„ì„í•  ì˜ìƒì„ ì—…ë¡œë“œí•´ì£¼ì„¸ìš”...")
uploaded = files.upload()
if not uploaded:
    print("âŒ ì˜ìƒì´ ì—…ë¡œë“œë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤!")
    exit()

video_path = list(uploaded.keys())[0]
print(f"ğŸ¬ ì—…ë¡œë“œëœ ì˜ìƒ: {video_path}")

# ğŸ¥ ë¹„ë””ì˜¤ ìº¡ì²˜
cap = cv2.VideoCapture(video_path)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps    = cap.get(cv2.CAP_PROP_FPS)

# ğŸ’¾ ì¶œë ¥ ë¹„ë””ì˜¤ ì €ì¥ ì„¤ì •
output_path = 'output_yolov12.mp4'
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

frame_count = 0
print("\nğŸš€ YOLOv12 ê°ì²´ ì¸ì‹ ì‹œì‘...")

# ğŸ”„ í”„ë ˆì„ ë°˜ë³µ ì²˜ë¦¬
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_count += 1
    print(f"\rğŸ“· í”„ë ˆì„ {frame_count} ì²˜ë¦¬ ì¤‘...", end='')

    # YOLOv12 ì¶”ë¡ 
    results = model(frame, verbose=False)[0]

    # ê²°ê³¼ ì‹œê°í™” (ì¸ì‹ëœ ê°ì²´ê°€ ê·¸ë ¤ì§„ í”„ë ˆì„)
    annotated_frame = results.plot()

    # í•´ìƒë„ ë¶ˆì¼ì¹˜ ë°©ì§€ìš© ë¦¬ì‚¬ì´ì¦ˆ
    if annotated_frame.shape[1] != width or annotated_frame.shape[0] != height:
        annotated_frame = cv2.resize(annotated_frame, (width, height))

    # ì¶œë ¥ í”„ë ˆì„ ì €ì¥
    out.write(annotated_frame)

# ğŸ‰ ë§ˆë¬´ë¦¬
cap.release()
out.release()

print("\nâœ… ì˜ìƒ ë¶„ì„ ì™„ë£Œ!")
print(f"ğŸ’¾ ê²°ê³¼ ì €ì¥: {output_path}")

# ğŸ•’ íŒŒì¼ ì €ì¥ ì§€ì—° ë°©ì§€
time.sleep(2)

# ğŸ“¥ ê²°ê³¼ ì˜ìƒ ë‹¤ìš´ë¡œë“œ
files.download(output_path)
```
