강사님이 주신 데이터 로드 및 시각화 코드 -> 오류 뜸

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Lidar/ Raser 데이터 로드
lidar_data = np.loadtxt('scene_lidar_data.csv', delimiter=',')
radar_data = np.loadtxt('scene_radar_data.csv', delimiter=',')

# 2. Lidar 포인트 시각화 (x, y 좌표 사용)
plt.scatter(lidar_data[:, 0], lidar_data[:, 1], s=1, label='Lidar')

# 3. Radar 탐지 객체 시각화 (극좌표 -> 직교좌표 변환)
radar_x = radar_data[:, 3] * np.cos(radar_data[:, 1])
radar_y = radar_data[:, 3] * np.sin(radar_data[:, 1])
plt.scatter(radar_x, radar_y, s=100, c='red', marker='x', label='Radar')

plt.xlabel('X (m)'); plt.ylabel('Y (m)')
plt.legend(); plt.grid(True); plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 예제용 Lidar 데이터 생성 (1000개 포인트)
# Lidar는 보통 주변 지형을 스캔한 포인트 클라우드라고 생각
lidar_data = np.random.randn(1000, 2) * 10  # x, y 좌표 분포

# 2. 예제용 Radar 데이터 생성 (20개 객체)
# Radar는 방향(angle)과 거리(range)를 측정함
num_radar_points = 20
angles = np.linspace(-np.pi, np.pi, num_radar_points)  # -180° ~ 180°
ranges = np.random.uniform(5, 20, size=num_radar_points)  # 거리 5m~20m

# Radar 데이터 구성: 열 4개짜리 배열 중 2번째(각도), 4번째(거리)만 사용
radar_data = np.zeros((num_radar_points, 4))
radar_data[:, 1] = angles
radar_data[:, 3] = ranges

# 3. Lidar 시각화
plt.scatter(lidar_data[:, 0], lidar_data[:, 1], s=1, label='Lidar')

# 4. Radar 시각화 (극좌표 → 직교좌표)
radar_x = radar_data[:, 3] * np.cos(radar_data[:, 1])
radar_y = radar_data[:, 3] * np.sin(radar_data[:, 1])
plt.scatter(radar_x, radar_y, s=100, c='red', marker='x', label='Radar')

# 5. 그래프 꾸미기
plt.xlabel('X (m)')
plt.ylabel('Y (m)')
plt.legend()
plt.grid(True)
plt.axis('equal')  # 축 비율 1:1
plt.title('Simulated Lidar & Radar Visualization')
plt.show()
```
