```python
# 복잡한 코드 이 코드는 센서 기반 자율주행 자동차 시뮬레이션 코드
# 1. 센서 초기화 함수
def init_sensors():
    """센서 기본값 설정"""
    sensors = {
        "앞": 15.0,      # 기본값
        "왼쪽": 15.0,
        "오른쪽": 15.0
    }
    print("10 - 센서 초기화 완료", sensors)
    return sensors

# 2. 앞쪽 스캔 함수
def scan_front(x, y, obstacles, sensors):
    """앞쪽 장애물 스캔"""
    print("16- 앞쪽 스캔 시작")
    if (x, y+1) in obstacles:
        print("17", x, y, "앞쪽 1칸 장애물 발견")
        sensors["앞"] = 1.0
    elif (x, y+2) in obstacles:
        sensors["앞"] = 2.0
        print("22", x, y, "앞쪽 2칸 장애물 발견")
    elif (x, y+3) in obstacles:
        sensors["앞"] = 3.0
        print("25", x, y, "앞쪽 3칸 장애물 발견")
    else:
        print("27", x, y, "앞쪽 장애물 없음")
    print("28 - 앞쪽 스캔 완료")

# 3. 왼쪽 스캔 함수
def scan_left(x, y, obstacles, sensors):
    """왼쪽 장애물 스캔"""
    print("33- 왼쪽 스캔 시작")
    if (x-1, y) in obstacles:
        sensors["왼쪽"] = 1.0
        print("36", x, y, "왼쪽 1칸 장애물 발견")
    elif (x-2, y) in obstacles:
        sensors["왼쪽"] = 2.0
        print("39", x, y, "왼쪽 2칸 장애물 발견")
    elif (x-3, y) in obstacles:
        sensors["왼쪽"] = 3.0
        print("42", x, y, "왼쪽 3칸 장애물 발견")
    else:
        print("44", x, y, "왼쪽 장애물 없음")
    print("45 - 왼쪽 스캔 완료")

# 4. 오른쪽 스캔 함수
def scan_right(x, y, obstacles, sensors):
    """오른쪽 장애물 스캔"""
    print("50 - 오른쪽 스캔 시작")
    if (x+1, y) in obstacles:
        sensors["오른쪽"] = 1.0
        print("53", x, y, "오른쪽 1칸 장애물 발견")
    elif (x+2, y) in obstacles:
        sensors["오른쪽"] = 2.0
        print("56", x, y, "오른쪽 2칸 장애물 발견")
    elif (x+3, y) in obstacles:
        sensors["오른쪽"] = 3.0
        print("59", x, y, "오른쪽 3칸 장애물 발견")
    else:
        print("61", x, y, "오른쪽 장애물 없음")
    print("62 - 오른쪽 스캔 완료")

# 5. 센서 데이터 업데이트 함수 (디버그 print 포함)
def update_sensor_data(position, obstacles):
    """현재 위치에서 센서 데이터 업데이트"""
    print("67 - update_sensor_data 함수 시작")
    x, y = position["x"], position["y"]
    print("69- 위치 추출 완료")
    print("70", x, y)

    # 센서 초기화
    sensors = init_sensors()

    # 각 방향별 스캔 실행
    scan_front(x, y, obstacles, sensors)
    scan_left(x, y, obstacles, sensors)
    scan_right(x, y, obstacles, sensors)

    print("80 - 센서 업데이트 완료", sensors)
    return sensors

# 6. 장애물 감지 함수
def detect_obstacles(sensors, safe_distance):
    """장애물을 감지하는 함수 (원본 그대로)"""
    print("86 - detect_obstacles 함수 시작")
    print("87", sensors, safe_distance)
    obstacles = []

    for direction, distance in sensors.items():
        print("91", direction, distance, "체크 중")
        if distance < safe_distance:
            obstacles.append(direction)
            print("94", direction, "위험! 장애물 목록에 추가")
        else:
            print("96", direction, "안전")

    print("98 - 장애물 감지 완료", obstacles)
    return obstacles

# 7. 행동 결정 함수
def decide_action(obstacles):
    """장애물 정보를 바탕으로 행동을 결정하는 함수 (원본 그대로)"""
    print("104 - decide_action 함수 시작")
    print("105", obstacles, "입력된 장애물 목록")

    if len(obstacles) == 0: # len()은 Python의 내장 함수로, 객체의 길이(length), obstacles 리스트(또는 다른 컬렉션)가 비어있는지 확인하는 조건문
        action = "직진"
        print("109", "장애물 없음 -> 직진")
    elif "앞" in obstacles:
        print("111", "앞에 장애물 있음, 좌우 확인")
        if "왼쪽" not in obstacles:
            action = "좌회전"
            print("114", "왼쪽 안전 -> 좌회전")
        elif "오른쪽" not in obstacles:
            action = "우회전"
            print("117", "오른쪽 안전 -> 우회전")
        else:
            action = "정지"
            print("120", "모든 방향 위험 -> 정지")
    else:
        action = "주의해서 직진"
        print("123", "앞은 안전, 옆만 위험 -> 주의해서 직진")

    print("125 - 행동 결정 완료", action)
    return action

# 8. 간단한 자율주행 자동차 클래스 (디버그 포함)
class SimpleCar:
    def __init__(self, name, x, y):
        """자동차 생성"""
        print("132 - SimpleCar __init__ 시작")
        self.name = name
        self.position = {"x": x, "y": y}
        self.safe_distance = 10.0
        print("136", name, x, y, "자동차 생성 완료")

    def move_forward(self):
        """전진"""
        print("140 - move_forward 시작")
        self.position["y"] += 1
        print("142", self.position["x"], self.position["y"], "전진 완료")
        print(f"🚗 {self.name}이(가) 전진합니다. 위치: ({self.position['x']}, {self.position['y']})")

    def turn_left(self):
        """좌회전"""
        print("147 - turn_left 시작")
        self.position["x"] -= 1
        print("149", self.position["x"], self.position["y"], "좌회전 완료")
        print(f"↖️ {self.name}이(가) 좌회전합니다. 위치: ({self.position['x']}, {self.position['y']})")

    def turn_right(self):
        """우회전"""
        print("154 - turn_right 시작")
        self.position["x"] += 1
        print("156", self.position["x"], self.position["y"], "우회전 완료")
        print(f"↗️ {self.name}이(가) 우회전합니다. 위치: ({self.position['x']}, {self.position['y']})")

    def stop(self):
        """정지"""
        print("161 - stop 시작")
        print("162", self.position["x"], self.position["y"], "정지")
        print(f"🛑 {self.name}이(가) 정지합니다. 위치: ({self.position['x']}, {self.position['y']})")

    def execute_action(self, action):
        """행동 실행"""
        print("167 - execute_action 시작")
        print("168", action, "실행할 행동")

        if action == "직진" or action == "주의해서 직진":
            print("171", "직진 관련 행동 -> move_forward 호출")
            self.move_forward()
        elif action == "좌회전":
            print("174", "좌회전 -> turn_left 호출")
            self.turn_left()
        elif action == "우회전":
            print("177", "우회전 -> turn_right 호출")
            self.turn_right()
        elif action == "정지":
            print("180", "정지 -> stop 호출")
            self.stop()

        print("183 - execute_action 완료")

    def run_simulation(self, obstacles):
        """자율주행 시뮬레이션"""
        print("187 - run_simulation 시작")
        steps = 0
        max_steps = 10

        print("🚗 자율주행 장애물 감지 시스템")
        print("-" * 30)
        print(f"🚧 장애물 위치: {list(obstacles)}")
        print("194", obstacles, "시뮬레이션 시작")

        while steps < max_steps:
            print(f"\n--- 스텝 {steps + 1} ---")
            print("198", steps + 1, "스텝 시작")
            print(f"📍 현재 위치: ({self.position['x']}, {self.position['y']})")
            print("200", self.position["x"], self.position["y"], "현재 위치")

            # 센서 데이터 업데이트
            print("203 - 센서 업데이트 호출")
            sensors = update_sensor_data(self.position, obstacles)
            print(f"📡 센서 데이터: {sensors}")
            print("20", sensors, "센서 데이터 받음")

            # 장애물 감지 (원본 함수 그대로 사용)
            print("2 - 장애물 감지 호출")
            detected_obstacles = detect_obstacles(sensors, self.safe_distance)
            print(f"🚨 감지된 장애물: {detected_obstacles}")
            print("212", detected_obstacles, "감지된 장애물")

            # 행동 결정 (원본 함수 그대로 사용)
            print("215 - 행동 결정 호출")
            action = decide_action(detected_obstacles)
            print(f"🎯 결정된 행동: {action}")
            print("218", action, "결정된 행동")

            # 행동 실행
            print("221 - 행동 실행 호출")
            self.execute_action(action)

            # 정지했으면 종료
            if action == "정지":
                print("226", "정지 조건 만족, 시뮬레이션 종료")
                print("⚠️ 모든 방향 막혀서 시뮬레이션 종료!")
                break

            steps += 1
            print("231", steps, "스텝 완료")

        print(f"\n🏁 시뮬레이션 완료!")
        print("234 - run_simulation 완료")

# 9. 시뮬레이션 실행
if __name__ == "__main__":
    print("238- 메인 실행 시작")

    # 자동차 생성
    print("241- 자동차 생성")
    car = SimpleCar("현대차", 3, 0)

    # 장애물 설정
    obstacles = {(3, 2), (3, 3), (2, 4), (4, 5)}
    print("247", obstacles, "장애물 설정 완료")

    # 시뮬레이션 실행
    car.run_simulation(obstacles)
```
